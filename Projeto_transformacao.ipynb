# Precificação Inteligente
# Trata-se de um algoritmo capaz de analisar as características de um imóvel — como comodidade, tamanho, ocupação do espaço em determinado período — e sugerir à pessoa anfitriã um preço a ser cobrado por diárias que garanta o lucro em momentos de alta demanda.

```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

# Aplicando normalização dos dados

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

Desagrupando as listas


```python
colunas = list(dados.columns)
colunas
```

Utilizando método explode

```python
dados = dados.explode(colunas[3:])
dados
```

Resetando os índices para ordenar os dados

```python
dados.reset_index(inplace = True, drop= True)
dados.head()
```

Verificando os tipos de dados

```python
dados.info()
```

Convertendo dados numéricos

```python
import numpy as np
```

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

Convertendo dados numéricos com mais de uma coluna

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

Convertendo para o tipo float / Removendo caracteres especiais

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].map(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

Analise textual / tokenização simples

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

```python
dados['descricao_local'][3169]
```

REGEX

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex = True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\W)-(?!\W)', ' ',regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex = True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex = True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\W)-(?!\W)', ' ',regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```

# Dados Temporais

```python
dt_data = pd.read_json('/content/moveis_disponiveis.json')
dt_data.head()
```

```python
dt_data.info()
```

Transformando o dado em datetime

```python
dt_data['data'] = pd.to_datetime(dt_data['data'])
```

```python
dt_data.info()
```

Manipulando dados temporais

```python
dt_data.head()
```

```python
dt_data['data'].dt.strftime('%Y-%m')
```

Agrupando dados

```python
# Agrupando dados de cada mês e somando as vagas disponiveis que são iguais a True
subset = dt_data.groupby(dt_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset
```

Transformando dados na coluna preço

```python
dt_data.head()
```

```python
dt_data.info()
```

```python
# utilizamos o método fillna para preencher os elementos vazios por '0.0'
# definimos o parâmetro de inplace para True para substituir no DataFrame
dt_data['preco'].fillna('0.0', inplace = True)
```

```python
# apagamos o $ e as vírgulas com apply lambda
dt_data['preco'] = dt_data['preco'].apply(lambda x: x.replace('$', '').replace(',', ''))
```

```python
# transformamos os tipos de dados para float64
dt_data['preco'] = dt_data['preco'].astype(np.float64)
dt_data
```
